```python
images = np.array(images)
np.save("image",images)
np.save("labels", labels)

image = np.load("image.npy",allow_pickle = True)
labels = np.load("labels.npy",allow_pickle = True)

img_shape = np.arange(image.shape[0])
np.random.shuffle(img_shape)# se toma un rango para moder mezclar aleatoriamente y tener un mejor medelo
image = image[img_shape]
labels = labels[img_shape]
```

```python
num_classes = len(np.unique(labels))
len_data = len(image)
```

```python
x_train,x_test = image[(int)(0.1*len_data):],image[:(int)(0.1*len_data)]
y_train,y_test = labels[(int)(0.1*len_data):],labels[:(int)(0.1*len_data)]

y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

```

```python
l2_reg = 0.001


cnn_model = Sequential()
cnn_model.add (Conv2D(filters = 32, kernel_size = (2,2), input_shape = (50,50, 3),
                      activation = 'relu', kernel_regularizer = l2(l2_reg)))
cnn_model.add (MaxPool2D(pool_size = (2,2)))
cnn_model.add (Conv2D(filters = 64, kernel_size = (2,2),
                       activation = 'relu', kernel_regularizer = l2(l2_reg)))
cnn_model.add (MaxPool2D (pool_size = (2,2)))
cnn_model.add (Conv2D(filters = 128, kernel_size = (2,2),
                      activation = 'relu', kernel_regularizer = l2(l2_reg)))
cnn_model.add (MaxPool2D(pool_size = (2,2)))
cnn_model.add (Dropout (0.1))


cnn_model.add (Flatten())


cnn_model.add (Dense (64, activation = 'relu'))#capa oculta de activacion relu
cnn_model.add (Dense (64, activation = 'relu'))#capa oculta de activacion relu
cnn_model.add(Dense(6, activation='softmax')) #capa de salida de 32 nueronas con activacion softmax

#CNN Model Summary
cnn_model.summary ()
```

```python
from tensorflow.keras.optimizers import Adam

opt = Adam(lr=0.001)

cnn_model.compile(loss='categorical_crossentropy',
                  optimizer=opt,
                  metrics=['accuracy'])
```

```python
import tensorflow as tf

class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self,epoch,logs={}):
    if (logs.get('accuracy')>0.98):
      print("\nAlcanzo 98% de precicion, se detiene el entrenamiento!")
      self.model.stop_training = True

callbacks = myCallback()
```

```python
history = cnn_model.fit(
    x_train, y_train,
    epochs=10,
    batch_size=32,
    validation_data=(x_test, y_test)
)




```

```python
scores = cnn_model.evaluate(x_test, y_test, verbose = 1)
print('Test loss:', scores [0])
print('Test accuracy:', scores [1]) #se clasifica el nivel del pedida del 98%
```

```python
figure = plt.figure(figsize = (10,5))
ax = figure.add_subplot (121)
ax.set_ylabel('% Exactitud')
ax.set_xlabel('# Epocas')
ax.plot(history.history['accuracy'])
ax.plot(history.history['val_accuracy'])
ax. legend(['Training Accuracy'
, 'Val Accuracy'])
bx = figure.add_subplot (122)
bx.set_ylabel('% Perdida')
bx.set_xlabel('# Epocas')
bx.plot(history.history['loss'])
bx.plot(history.history['val_loss'])
bx. legend(['Training Loss', 'Val Loss']) #graficamos las pedidadas de los 2 conjuntos
```
